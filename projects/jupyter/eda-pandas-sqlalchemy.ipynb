```python
from sqlalchemy.engine.url import URL
from sqlalchemy import create_engine, MetaData, Table, dialects
import pandas as pd
postgres_db = {'drivername': 'postgresql',
               'username': 'postgres',
               'password': 'postgres',
               'host': '172.17.0.1', # alternatively, 127.0.0.1/localhost if running standalone
               'port': 5432,
               'database': 'postgres'
               }
pgURL = URL(**postgres_db)
engine = create_engine(pgURL)
conn = engine.connect()
```

```python
query = """
SELECT * FROM information_schema.tables
"""
df = pd.read_sql(query, conn)
df
```

```python
query = """
SELECT * FROM information_schema.tables
WHERE table_schema = 'pg_catalog'
"""
df_2 = pd.read_sql(query, conn)
df_2
```
